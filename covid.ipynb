```python
# Convert wide format to long format
confirmed_long = confirmed_df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                   var_name='Date', value_name='Confirmed')
recovered_long = recovered_df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                   var_name='Date', value_name='Recovered')

deaths_long = deaths_df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                              var_name='Date', value_name='Deaths')
```

```python
# Merge on common columns
merged_long = confirmed_long.merge(recovered_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
merged_long = merged_long.merge(deaths_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
merged_long
```

```python
# Calculate active cases
merged_long['Active'] = merged_long['Confirmed'] - merged_long['Recovered'] - merged_long['Deaths']

merged_long.head()
```

## THERE IS AN INCREASING TREND AND A MONTHLY SEASONAL COMPONENT AND THE RESIDUAL SHOWS A PERIODIC PATTERN HENCE WE NEED TO ELIMINATE THESE TO GET A STATIONARY DATA

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# First difference
df_daily['First_Diff'] = df_daily['Confirmed'].diff().dropna()

# Plot differenced series
plt.figure(figsize=(12, 6))
plt.plot(df_daily['First_Diff'], label='First Difference of Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Differenced Confirmed Cases')
plt.title('First Difference of Time Series')
plt.legend()
plt.show()

# Check for stationarity
result = adfuller(df_daily['First_Diff'].dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

## SINCE ADF PROBABILITY VALUE IS STILL LARGER THAN 0.05 WE NEED FURTHER DIFFERENCING TO ACHIEVE STATIONARITY

```python
# Second difference
df_daily['Second_Diff'] = df_daily['First_Diff'].diff().dropna()

# Plot second differenced series
plt.figure(figsize=(12, 6))
plt.plot(df_daily['Second_Diff'], label='Second Difference of Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Second Differenced Confirmed Cases')
plt.title('Second Difference of Time Series')
plt.legend()
plt.show()

# Check for stationarity again
result = adfuller(df_daily['Second_Diff'].dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
df_daily['Log_Confirmed'] = np.log(df_daily['Confirmed'])
df_daily['Log_First_Diff'] = df_daily['Log_Confirmed'].diff().dropna()

# Plot log differenced series
plt.figure(figsize=(12, 6))
plt.plot(df_daily['Log_First_Diff'], label='Log First Difference of Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Log Differenced Confirmed Cases')
plt.title('Log First Difference of Time Series')
plt.legend()
plt.show()

# Check for stationarity
result = adfuller(df_daily['Log_First_Diff'].dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```
